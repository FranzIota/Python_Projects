```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(5)
```

```python
df.index.name = 'job_index'
```

## Index Reset

```python
df_indo = df[df['job_country'] == 'Indonesia'] 
df_indo.head(5)

# bisa diliat di df_indo kalau job index nya random 
```

```python
# Create new index 
# ini ngereset / delete kolom yang tadinya jadi index 
df_indo.reset_index(inplace=True)

```

```python
# new index ada di bagian paling kiri 
df_indo
```

## Set Index

```python
df_indo.set_index('job_index', inplace=True)
```

```python
df_indo
```

## Sort Index

```python
median_pivot = df_indo.pivot_table(values ='salary_year_avg', index='job_title_short', aggfunc= ['median', 'min', 'max'])
median_pivot
```

```python
# by= nya dibikin tuple () karena dia ada 2 baris yaitu salary n agg func
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)

# median_pivot.sort_values(by=('job_title_short', 'salary_year_avg'), ascending=False)
```

```python
# ini sorting dari job title nya 
median_pivot.sort_index(inplace=True)
```
