```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# return random data
df.sample(5, random_state=3)
```

1. Copy digunakan untuk membuat salinan dari dataframe
2. kalau cuma seperti: df_new = df, maka jika pada df_new ada kolom / baris yang dimanipulasi, maka data di df juga akan termanipulasi
3. dikarekan jika memakai: df_new = df, maka dataframe nya adalah sama, yang memiliki id yang sama sehingga otomatis terupdate.
4. dengan memakai .copy(), maka akan terbentuk salinan dengan id berbeda, yang mana df asli tidak akan berubah datanya.

```python
# copy
df_copy = df.copy()

print(id(df_copy))
print(id(df))
print(id(df_copy) == id(df))
```
