Goals:
1. Convert all NaN data into median values of each of those column
2. Delete all duplicate rows

```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.loc[:10, 'salary_year_avg':'salary_hour_avg']

# Nilai yang Nan, mau di convert jadi nilai median
```

```python
salary_yearly_median = df.salary_year_avg.median()
```

```python
salary_hourly_median = df.salary_hour_avg.median()
```

```python
# create dataframe baru, agar yang original tetap intact
df_filled = df

# reassign yang nan value dengan median value
df_filled['salary_year_avg'] = df_filled.salary_year_avg.fillna(salary_yearly_median)
df_filled['salary_hour_avg'] = df_filled.salary_hour_avg.fillna(salary_hourly_median)
```

```python
df_filled.loc[:10, 'salary_year_avg':'salary_hour_avg']
```

```python
# delete duplicate 
df_unique = df_filled

df_unique = df_unique.drop_duplicates()
print('Length of Original data  :     ', len(df_filled))
print('Length of droped data    :     ', len(df_unique))
print('Droped data              :     ', len(df_filled)-len(df_unique))
```

```python
# delete duplicate spesific (job title and company name)
df_unique = df_filled

df_unique = df_unique.drop_duplicates(subset=['job_title','company_name'])
print('Length of Original data  :     ', len(df_filled))
print('Length of droped data    :     ', len(df_unique))
print('Droped data              :     ', len(df_filled)-len(df_unique))
```
